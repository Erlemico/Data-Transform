```python
import pandas as pd
from IPython.display import display, HTML
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, roc_curve, auc
import seaborn as sns
import matplotlib.pyplot as plt

# File path
dataset_fortex = '/Users/Erlemico/Downloads/dataset-fortex-2023.csv'

# Set chunk size for processing large dataset
chunk_size = 10000

# Optionally, set dtype for column 15 (if you know its type, such as string)
dtype_dict = {15: str}  # Set column 15 to string type

# Read dataset in chunks, handle mixed types warning by setting low_memory=False
chunk_iter = pd.read_csv(dataset_fortex, chunksize=chunk_size, low_memory=False, dtype=dtype_dict)

# Processing chunks
for chunk in chunk_iter:
    # Display chunk as HTML (only first chunk for demonstration)
    display(HTML(chunk.head().to_html()))
    break  # Only show the first chunk for now

# If you want to read the entire dataset at once (instead of using chunking):
# df = pd.read_csv(file_path, low_memory=False, dtype=dtype_dict)
# display(HTML(df.to_html()))  # Uncomment this if you want to read all data at once
```
